### Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from scipy import stats
from sklearn.metrics import accuracy_score
```

### Loading the dataset

```python
dataset =pd.read_csv('Iris.csv')
```

```python
dataset
```

### Preprocessing the data

```python
dataset.isnull().any()
```

```python
df= dataset.drop('Id',axis=1)
df
```

```python
df['Species'].value_counts()
```

### Label encoding for categorical data

```python
lb= LabelEncoder()
```

```python
df.iloc[:,4]= lb.fit_transform(df.iloc[:,4])
df
```

### Removing outliers

```python
z= np.abs(stats.zscore(df))
z
```

```python
threshold=3  # Define a threshold to identify an outlier.
np.where(z>threshold)
```

```python
df=df[(z<=3).all(axis=1)]
df
```

### Splitting into x, y

```python
x= df.iloc[:,0:3].values
x
```

```python
y= df.iloc[:,-1].values
y
```

### Splitting into train and test set

```python
from sklearn.model_selection import train_test_split  #previously cros_validation was used in sklearn
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

### Model Fitting

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt= DecisionTreeClassifier(criterion='gini',random_state=0,max_depth=5)
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
y_pred
```

```python
y_test
```

### Accuracy

```python
accuracy_score(y_test,y_pred)
```

### Creating Decision Tree

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dt, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
